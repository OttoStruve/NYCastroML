<p class='h2'>SDSS</p>

- Photometric and spectroscopic surveys on the same telescope
- Main survey (SDSS-I and SDSS-II) completed just a few years ago
- Stats below are from Data Release 7 (DR7)

<p class='h4'>Photometry</p>

- Imaging data was taken in _stripes_ (see image below)
- 5 photometric bands, in increasing wavelength, $u,g,r,i,z$
- ~400 million objects detected over ~1/3 of the sky
- Stripe 82 -- along the celestial equator -- re-imaged multiple times, provides time-domain information

```python
from IPython.display import Image
Image("http://www.sdss3.org/images/dr8_spectro_coverage.png", width=800)
```

- The full imaging data is many 10's of terabytes -- for the purposes of the examples in this book, they limit the focus to a small subset of the full footprint.
- This includes ~330,000 sources, (mostly) cleaned of bad data

```python
sdss_data = datasets.fetch_imaging_sample()
```

```python
figure(figsize=(4,4))
plot(sdss_data['ra'], sdss_data['dec'],
     marker='.', linestyle='none', alpha=0.1)
xlim(360, 0)
ylim(-90, 90)
```

```python
figure(figsize=(12,6))
plot(sdss_data['ra'], sdss_data['dec'],
     marker='.', linestyle='none', alpha=0.05)
title("Source density")
xlim(10.5, -0.5)
ylim(-2, 2)
```

Now, as an example, we separate the stars and galaxies and plot a color-magnitude diagram:

```python
stars = sdss_data[sdss_data['type'] == 6]
galaxies = sdss_data[sdss_data['type'] == 3]
stars.size, galaxies.size
```

```python
fig,axes = subplots(1,2,figsize=(12,4),
                    sharex=True,sharey=True)

axes[0].plot(stars['gRaw']-stars['rRaw'], 
             stars['rRaw']-stars['iRaw'],
             alpha=0.02, marker='.', linestyle='none')
axes[0].set_xlabel(r"$g-r$")
axes[0].set_ylabel(r"$r-i$")
axes[0].set_title("Stars")

axes[1].plot(galaxies['gRaw']-galaxies['rRaw'], 
             galaxies['rRaw']-galaxies['iRaw'],
             alpha=0.02, marker='.', linestyle='none')
axes[1].set_xlabel(r"$g-r$")
axes[1].set_title("Galaxies")

axes[0].set_xlim(-0.5, 3)
axes[0].set_ylim(-1, 2)
```

<p class='h4'>Spectroscopy</p>

- Moderate resolution spectroscopy from 3800-9200$A$
     - ~2 million spectra
- They provide an interface for downloading SDSS spectra:

```python
spectrum = datasets.fetch_sdss_spectrum(2872, 54533, 71)
```

```python
from scipy.ndimage import gaussian_filter

figure(figsize=(12,6))
plot(spectrum.wavelength(), gaussian_filter(spectrum.spectrum,2), 
     drawstyle='steps', alpha=0.75, color='k')
xlim(min(spectrum.wavelength()), max(spectrum.wavelength()))
xlabel(r"Wavelength [$\AA$]")
ylabel(r"Flux")
```
