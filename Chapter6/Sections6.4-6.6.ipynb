## 6.4.2 : K-Means (sum of squares minimization)

```python
import colors # My personal library of colors from colorbrewer2

#colors.qual_10.show()
Kcolors = colors.qual_10.tuples
```

Assign points to a group based on their closeness to centroids $\mu_k$ (which I will initially chose at random)

```python
def distance(x1, x2):
    return np.sqrt( np.sum((x1-x2)**2) )
```

```python
K = 10
def assign_groups(data, centroids, k=K):
    groups = [[] for i in range(K)]
    for x in data:  #<--- This is slow, could probably be optimized
        dist = np.array([distance(x,mu) for mu in centroids])
        igroup = np.where(dist == min(dist))[0]
        groups[igroup[0]].append([x[0],x[1]])
    return groups
```

```python
def plot_groups(groups, centroids):
    plt.ylim(xmin, xmax)
    plt.ylabel('x (Mpc)')
    plt.xlim(ymin, ymax)
    plt.xlabel('y (Mpc)')
    for i in range(K):
        grp_array = np.array(groups[i])
        plt.scatter(grp_array[:,1], grp_array[:,0], s=2, lw=0, c=Kcolors[i])
        plt.plot(centroids[i][1], centroids[i][0], marker='o', ls='', c=Kcolors[i])
    return
```

```python
mu_init = np.array([np.random.uniform(size=K,low=xmin,high=xmax), 
           np.random.uniform(size=K,low=ymin,high=ymax)]).transpose()
g_init  = assign_groups(X, mu_init)
plot_groups(g_init, mu_init)
```

Now adjust the centroids based on the current clusters, then re-assign clusters.

```python
def new_centroids(data, groups):
    return np.array([np.mean(G,axis=0) for G in groups])
```

```python
NSTEPS = 20
def kmeans(data, groups, n=NSTEPS):
    for i in range(n):
        next_mu = new_centroids(data, groups)
        next_group = assign_groups(data, next_mu)
    
    return next_group, next_mu
```

```python
g_final, mu_final = kmeans(X, g_init)
plot_groups(g_final, mu_final)
```

## 6.4.4 : Mean Shift (Nonparametric Density Estimation)

+ Use a nonparametric density estimate (Section 6.1)

+ Move cluster center in the direction of $\log$ gradient

+ Cluster centers will stop moving when it reaches a peak

To produce the $m+1$-th step:

$x_i^{m+1} = x_i^m +$ step size ($a$) $\times$ slope of log kernel density estimation ($\nabla \log \hat{f}^m(x_i^m)$)

$$ = x_i^m + \frac{a}{\hat{f}^m(x_i^m)} \nabla \hat{f}^m(x_i^m)$$

```python
ISEED    = 42
center0 = X[ISEED]

ax = plt.subplot(1,1,1)
plot_GreatWall(ax)
plt.plot(center0[1], center0[0], marker='o', color='r')
```

In terms of angular distances, the two point correlation function is the probability of finding another point (galaxy) a distance $\theta$ away.

$$ dP_{12} = \rho^2 (1 + \omega(\theta)) $$
where $\rho$ is the density of points (i.e. galaxies per square degree)

OR you can think of it as a dot product between the "density contrast" $(\rho - \bar{\rho})/\rho$ at two different points ($x$ and $x+\theta$):
$$ \omega(\theta) = \left< \frac{\delta \rho(x)}{\rho}\ \frac{\delta \rho(x+\theta)}{\rho} \right> $$

**Long story short:**  If you an $n$-point ($n>2$) correlation function, it gets complicated and computationally intensive $O(N^n)$ very quickly!

# 6.6 : Which Density Estimation and Clustering Algorithms Should I Use?

Good question! The tool you use should depend on what you want to get out.  The AstroML authors define a few practical considerations.  **See Table 6.1 for more details.**

**Accuracy:**

+ Parametric models -- accuracy is poor if "the true underlying function is not a member of th echosen model class"

+ Non-parametric models -- do much better

+ winners: *Hierarchical clustering, Correlation functions, any of the density estimation methods*

+ losers: *K-means, Max-radius minimization*

**Interpretability, predictive power, or SCIENCE:**

+ An "accute" problem for clustering

+ Parametric models -- generally better because your parameters are understood (even if they're wrong)

+ Non-parametric models are often less clear

+ winners: *Mean shift, Kernel density estimation, extreme deconvolution*

+ losers: *Hierarchical clustering*

**Simplicity:**

+ How "fiddly" ("not-automatable", sensitive to initial conditions, value of cirtical parameters) is it?

+ The weak link for clustering methods: **K** (total number of clusters)

+ winners: *Hierarchical clustering* ??

+ losers: *K-means, max-radius minimization, K-nearest neighbor* ??

**(Computational) Speed:**

+ $O(N)$ and $O(N \log N)$ are generally considered great. If needed, can be reduced to $O(\log N)$ with tree-based algorithms.

+ $O(N^2)$ is slow, but can sometimes be reduced using tree-based algorithms.

+ $O(N^n)$ with $n \geq 3$ are bad!
