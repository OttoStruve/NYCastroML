I want to point out a few nice things about the distribution objects defined in Scipy. Let's import the normal distribution as an example.

```python
from scipy.stats import norm
```

As described in section 3.3, you can sample random variables:

```python
n = hist(norm.rvs(size=1000), bins=25)
```

But you can also _fit_ a distribution using the `.fit()` method (which does maximum likelihood estimation in the background):

```python
loc,scale = 2., 0.25
data = norm.rvs(loc=loc, scale=scale, size=100)

mle_loc,mle_scale = norm.fit(data)

n = hist(data, bins=25, normed=True, alpha=0.5)

x = np.linspace(1.,3,100)
plot(x, norm.pdf(x, loc=loc, scale=scale))
plot(x, norm.pdf(x, loc=mle_loc, scale=mle_scale))
xlim(1, 4)
```
