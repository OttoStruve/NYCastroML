```python
win1=np.where(((sp[0]>5050) & (sp[0]<5100)) | ((sp[0]>5500) & (sp[0]<5600)))
winall1=np.where(((sp[0]>5050) & (sp[0]<5600)))
```

```python
from scipy import stats 
slope1, intercept1, r_value, p_value, std_err = stats.linregress(sp[0][win1[0]],sp[1][win1[0]])
```

```python
pl.plot(sp[0][win1[0]],sp[1][win1[0]],'k')
pl.plot(sp[0][winall1[0]],sp[1][winall1[0]],'r')
pl.plot(np.arange(5050,5600), np.arange(5050,5600)*slope1+intercept1,'g')
pl.show()
```

```python
win2=np.where(((sp[0]>5700) & (sp[0]<5750)) | ((sp[0]>6100) & (sp[0]<6150)))
winall2=np.where(((sp[0]>5700) & (sp[0]<6150)))
slope2, intercept2, r_value, p_value, std_err = stats.linregress(sp[0][win2[0]],sp[1][win2[0]])
pl.plot(sp[0][win2[0]],sp[1][win2[0]])
pl.plot(sp[0][winall2[0]],sp[1][winall2[0]])
pl.plot(np.arange(5700,6150), np.arange(5700,6150)*slope2+intercept2)
pl.show()
```

```python
#renormalize and populate the caussian with counts.
feat1=(sp[1][winall1]-(sp[0][winall1]*slope1+intercept1))
tmp= (abs(feat1[1]-feat1[:-1]))
norm1=min(tmp[np.where(tmp > 0)[0]])
feat1=feat1/norm1
feat1=feat1-min(feat1)
pl.plot(sp[0][winall1],feat1)
pl.show()
```

```python
#renormalize and populate the caussian with counts.
feat2=(sp[1][winall2]-(sp[0][winall2]*slope2+intercept2))
tmp= (abs(feat2[1]-feat2[:-1]))
norm2=min(tmp[np.where(tmp > 0)[0]])
feat2=feat2/norm2
feat2=feat2-min(feat2)
pl.plot(sp[0][winall2],feat2)
pl.show()
```

```python
pop1=[winall1[0][0]]*feat1[0]
pop2=[winall2[0][0]]*feat2[0]
```

```python
for i,x in enumerate(winall1[0][1:]):
    pop1 = pop1+[x]*feat1[i]
pl.hist(pop1, bins=winall1[0], alpha=0.1)
pl.show()
```

```python
for i,x in enumerate(winall2[0][1:]):
    pop2 = pop2+[x]*feat2[i]
pl.hist(pop2, bins=winall2[0], alpha=0.1)
pl.show()
```

```python
lowest_bic = [np.infty,np.infty]
best_gmm=[0,0]
bic=[[],[]]
all_gmm=[[],[]]
X=[np.array(pop1),np.array(pop2)]#,np.array(sp[1][winall2]-sp[0][winall2]*slope2+intercept2)]
```

```python
for i,data in enumerate(X):
#i=0
    for n_components in range(1,10):
            # Fit a mixture of gaussians with EM
        gmm = mixture.GMM(n_components=n_components)
        gmm.fit(data)
        print gmm.bic
        bic[i].append(gmm.bic(data))
        all_gmm[i].append(gmm)
        if bic[i][-1] < lowest_bic[i]:
            lowest_bic[i] = bic[i][-1]
            best_gmm[i] = gmm
```
