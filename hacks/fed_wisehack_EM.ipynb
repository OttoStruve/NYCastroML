```python
N = np.arange(1, 8)
```

```python
models = []

for ng in N:
    gmm = GMM(ng, n_iter=100,covariance_type='diag')
    gmm.means_ = np.array(means)
    #print vars.diagonal()
    gmm.covars_ = vars.diagonal()
    #np.array([vars[0,0],cars[1,1],vars[2,2]]) ** 2
    gmm.weights_ = np.array([1, 1, 1])
    gmm.fit(allcol)
    models.append(gmm)
```

```python
print gmm.get_params
AIC = [m.aic(allcol) for m in models]
BIC = [m.bic(allcol) for m in models]
print AIC,BIC
pl.scatter(range(len(BIC)),BIC)
pl.scatter(range(len(AIC)),AIC)
    
```

```python
spl = pl.subplot(1, 1, 1)
bars = []
bbar = BIC
bars.append(pl.bar(1-0.2, (bbar/bbar[0]),width=.2, color='r', label='BIC'))
bhbar = AIC
bars.append(pl.bar(1+0.2, (hbar/hbar[0]),width=.2, color='b', label='AIC'))
for i in (0,N):
    xpos = np.arange(len(N))+1
    bbar = BIC
    bars.append(pl.bar(xpos-0.2, (bbar/bbar[0]),width=.2, color='r'))
    hbar = AIC
    bars.append(pl.bar(xpos+0.2, (hbar/hbar[0]),width=.2, color='b'))
pl.legend()
pl.yscale('log')
i_best = np.argmin(BIC)
gmm_best = models[i_best]
print "best fit converged:", gmm_best.converged_
print "BIC: n_components =  %i" % N[i_best]
print gmm_best.means_
```
