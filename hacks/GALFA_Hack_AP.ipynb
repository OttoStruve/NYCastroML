```python
from astropy.io import fits

filename = "GALFA_HI_RA+DEC_148.00+10.35_N_256_2x2.fits"
hdulist = fits.open(filename)

%matplotlib inline
```

```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.decomposition import NMF
```

```python
cube = hdulist[0].data
spectrum = cube[:,0,0]
```

```python
fig, ax = plt.subplots(figsize=(10,7.5))
ax.plot(spectrum, markersize=2, color='black')
```

```python
cv = cube.view()
cv2d = np.reshape(cv, (256, 256*256) )
cv2dr = cv2d.T
cv2dr = cv2dr.clip(min=0, max=None)
cv2dr.shape
```

```python
#Fit NMF components
nmf = NMF(n_components=5)
nmf.fit(cv2dr)
```

```python
proj = nmf.transform(cv2dr)
```

```python
components = proj.reshape(256,256,5)
```

```python
plt.imshow(components[:,:,1],interpolation='nearest',origin='lower')
```

```python
xr = yr = np.arange(256)
X,Y = np.meshgrid(xr,yr)
X = X.reshape(256*256)
Y = Y.reshape(256*256)
```

```python
posPlusSpec1 = np.vstack((X/256.0,Y/256.0,proj[:,1]/proj[:,1].max())).transpose()
posPlusSpec2 = np.vstack((X/256.0,Y/256.0,proj[:,0]/proj[:,0].max(),proj[:,1]/proj[:,1].max())).transpose()
```

```python
posPlusSpec1.shape
```

```python
#5 nearest neighbors in 2 dimensional embedding
k = 5
n = 2
```

```python
lle1 = LocallyLinearEmbedding(k,n)
lle2 = LocallyLinearEmbedding(k,n)
```

```python
lle1.fit(posPlusSpec1)
lle2.fit(posPlusSpec2)
```

```python
posPlusSpecProj1 = lle1.transform(posPlusSpec1)
posPlusSpecProj2 = lle2.transform(posPlusSpec2)
```

```python
posPlusSpecProj2.shape
```

```python
fig,ax = plt.subplots(1,2,figsize=(16,8))
ax[0].scatter(posPlusSpecProj1[:,0],posPlusSpecProj1[:,1],c=proj[:,0],edgecolor='none',marker='o')
ax[1].scatter(posPlusSpecProj2[:,0],posPlusSpecProj2[:,1],c=proj[:,0],edgecolor='none',marker='o')
```
