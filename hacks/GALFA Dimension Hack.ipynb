####Here's loading in a spectral line data cube of HI ISM from Arecibo, and decomposing it using NMF.

###Challenge: Can it be decomposed in a way consistent with what is seen as individual clouds by eye?

```python
import os
from astropy.io import fits

filename = "GALFA_HI_RA+DEC_148.00+10.35_N_256_2x2.fits"
if not os.path.exists(filename):
    ret = os.system("wget https://dl.dropboxusercontent.com/u/15321811/GALFA_HI_RA+DEC_148.00+10.35_N_256_2x2.fits -O {}".format(filename))

    if ret is not 0:
        raise IOError("Unable to retrieve file.")

hdulist = fits.open(filename)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from sklearn.decomposition import NMF
```

```python
cube = hdulist[0].data
```

```python
spectrum = cube[ : ,0, 0 ]
```

```python
fig, ax = plt.subplots(figsize=(10,7.5))
ax.plot(spectrum, markersize=2, color='black')
plt.show()
```

```python
cv = cube.view()
cvsub = cv.view()
cvsub = cvsub[100:150, 100:150, 100:150]
cv2dsub = np.reshape(cvsub, (50, 50*50) )
cv2drsub = cv2dsub.T
cv2d = np.reshape(cv, (256, 256*256) )
cv2dr = cv2d.T
cv2dr.shape
cv2drsub.shape
```

```python
spectrum2 = cv2dr[300, ]
fig, ax = plt.subplots(figsize=(10,7.5))
ax.plot(spectrum2, markersize=2, color='black')
plt.show()
```

```python
nmf = NMF(n_components=5)
```

```python
cv2drsub = cv2drsub.clip(min=0, max=None)
```

```python
nmf.fit(cv2drsub)
```

```python
proj = nmf.transform(cv2drsub)
```

```python
proj.shape
projrs = np.reshape(proj, [50, 50, 5])
```

```python
comp = nmf.components_
```

```python
im0 = (projrs[:,:,1])
plt.imshow(im0, interpolation='nearest', origin='lower')
plt.show()
```

```python
c0 = comp[1, ]
fig, ax = plt.subplots(figsize=(10,3.5))
ax.plot(c0, markersize=2, color='black')
plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding
k=5
n=2
lle = LocallyLinearEmbedding(k,n)
lle.fit(cv2drsub)
prj = lle.transform(cv2drsub)
```

```python
prj.shape
```

```python
fig, ax = plt.subplots(figsize=(10,7.5))
other = np.reshape(im0, 50*50)
ax.scatter(prj[:,0], prj[:,1], c=other, 
           marker='o', edgecolor='none', alpha=0.5)
```
