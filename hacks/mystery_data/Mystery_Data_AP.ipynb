```python
import numpy as np
import matplotlib.pyplot as plt
import triangle
```

```python
#Load the data from file
x,y,sigma_y = np.loadtxt("data.csv",delimiter=",").transpose()
```

```python
plt.errorbar(x,y,sigma_y,linestyle="none",marker="o")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

Likelihood function (from BAW, no copyright infringement intended):

$$\ln p_{\mathrm{good},i} = - \frac{1}{2} \left[ \frac{(y_i - (a x_i + b))^2}{\sigma_i^2} + \ln \left(2 \pi \sigma_i^2 \right) \right]$$

$$\ln p_{\mathrm{bad},i} = - \frac{1}{2} \left[ \frac{(y_i - (a x_i + b))^2}{\sigma_{\mathrm{bad}}^2} + \ln \left(2 \pi \sigma_{\mathrm{bad}}^2 \right) \right]$$

$$\ln p(X|\theta) = \sum_i \ln \left[ (1-Q_{\mathrm{bad}}) p_{\mathrm{good},i} + Q_{\mathrm{bad}} p_{\mathrm{bad},i} \right] $$

Priors:

$$p(a) = U(0,50)$$

$$p(b) = U(4,8)$$

$$p(\ln \sigma_{\mathrm{bad}}) = U(-2,2)$$

$$p(Q_{\mathrm{bad}}) = U(0,1)$$

```python
#Code up the above likelihood function
from scipy.misc import logsumexp

def logLikelihood(p,x,y,sigma_y):

	#Parameters are p=(a,b,Qbad,log(sigma_bad))
	a = p[0]
	b = p[1]
	Qbad = p[2]
	logSigmabad = p[3]

	#Uniform on a=U(0,50) ; b=U(4,8); Qbad=U(0,1) ; logSigmaBad=U(-2,2)
	if a<0. or a>50.:
		return -np.inf

	if b<4. or b>8.:
		return -np.inf

	if Qbad<0. or Qbad>1.:
		return -np.inf

	if logSigmabad<-2. or logSigmabad>2.:
		return -np.inf

	#Pgood
	logPgood = -0.5*(((y-a*x-b)/sigma_y)**2 + np.log(2*np.pi*sigma_y**2))
	#Pbad
	logPbad = -0.5*(((y-a*x-b)/np.exp(logSigmabad))**2 + np.log(2*np.pi) + 2*logSigmabad)

	#Data likelihood
	w = np.ones((2,len(x)))
	w[0] = 1.-Qbad
	w[1] = Qbad
	
	return logsumexp(np.vstack((logPgood,logPbad)),b=w,axis=0).sum()
```

```python
#Setup the MCMC sampler using emcee
import emcee

#4 parameters (a,b,Qbad,logSigmaBad)
ndim = 4
#32 walkers
nwalkers = 32
#Chain starts from here (initial positions of the walkers)
p0 = np.ones((nwalkers,ndim))
p0[:,0] = np.random.uniform(low=0.0,high=50.0,size=nwalkers)
p0[:,1] = np.random.uniform(low=4.0,high=8.0,size=nwalkers)
p0[:,2] = np.random.uniform(low=0.0,high=1.0,size=nwalkers)
p0[:,3] = np.random.uniform(low=-2.0,high=2.0,size=nwalkers)

#Sampler initialization
sampler = emcee.EnsembleSampler(nwalkers,ndim,logLikelihood,args=[x,y,sigma_y])
```

```python
#Burn in step
pos,prob,state = sampler.run_mcmc(p0,1000)
sampler.reset()
```

```python
#Run the chain after the burn in
sampler.run_mcmc(pos, 10000)
```

```python
#Plot the results using triangle
true_params = (0.5,6,0.5,0.0)
extents = [(0.5*truth,1.5*truth) for truth in true_params]
figure = triangle.corner(sampler.flatchain,labels=[r"$a$",r"$b$",r"$Q_{bad}$",r"$\log{\sigma_{bad}}$"],truths=true_params,plot_datapoints=False)
```
