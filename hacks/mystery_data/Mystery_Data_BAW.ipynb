# Mystery Data

We will be using [this data](https://gist.github.com/adrn/9908500).

```python
import numpy as np
import pylab
data = np.genfromtxt('./data.csv', delimiter=',',names=True)
p = pylab.errorbar(data['x'],data['y'],data['sigma_y'],fmt='ko')
```

Likelihood function:

$$\ln p_{\mathrm{good},i} = - \frac{1}{2} \left[ \frac{(y_i - (a x_i + b))^2}{\sigma_i^2} + \ln \left(2 \pi \sigma_i^2 \right) \right]$$

$$\ln p_{\mathrm{bad},i} = - \frac{1}{2} \left[ \frac{(y_i - (a x_i + b))^2}{\sigma_{\mathrm{bad}}^2} + \ln \left(2 \pi \sigma_{\mathrm{bad}}^2 \right) \right]$$

$$\ln p(X|\theta) = \sum_i \ln \left[ (1-Q_{\mathrm{bad}}) p_{\mathrm{good},i} + Q_{\mathrm{bad}} p_{\mathrm{bad},i} \right] $$

Priors:

$$p(a) = U(0,50)$$

$$p(b) = U(4,8)$$

$$p(\ln \sigma_{\mathrm{bad}}) = U(-2,2)$$

$$p(Q_{\mathrm{bad}}) = U(0,1)$$

```python
from scipy.misc import logsumexp
import emcee
import triangle
#
#
#
def line_model(theta,x):
    a, b, ln_s_bad, Q_bad = theta
    return a*x + b
#
#
#
def lnprior(theta):
    a, b, ln_s_bad, Q_bad = theta
    if a < 0 or a > 50:
        return -np.inf
    if b < 4 or b > 8:
        return -np.inf
    if ln_s_bad < -2 or ln_s_bad > 2:
        return -np.inf    
    if Q_bad < 0 or Q_bad > 1:
        return -np.inf
    return 0.
#
#
#
def lnlikelihood(theta,x,y,err):
    a, b, ln_s_bad, Q_bad = theta
    s_bad = np.exp(ln_s_bad)
    ln_p_good = -0.5*(((y-(a*x+b))/err)**2 + np.log(2.0*np.pi*err**2))
    ln_p_bad = -0.5*(((y-(a*x+b))/s_bad)**2 + np.log(2.0*np.pi) + 2.0*ln_s_bad)
    b = np.ones((2,len(x)))
    b[0] = 1. - Q_bad
    b[1] = Q_bad
    return logsumexp(np.vstack((ln_p_good, ln_p_bad)), b=b, axis=0)
#
#
#
def lnprob(theta,x,y,err):
    prior = lnprior(theta)
    if np.isinf(prior):
        return -np.inf
    return prior + np.sum(lnlikelihood(theta,x,y,err))
#
#
#
ndim, nwalkers = 4, 256
theta0 = np.zeros((nwalkers,ndim))
theta0[:,0] = np.random.uniform(0, 50, size=nwalkers) # a
theta0[:,1] = np.random.uniform(4, 8, size=nwalkers) # b
theta0[:,2] = np.random.uniform(-2, 2, size=nwalkers) # ln_s_bad
theta0[:,3] = np.random.uniform(0, 1, size=nwalkers) # Q_bad
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, 
                                args=(data['x'],data['y'],data['sigma_y']))
pos, prob, state = sampler.run_mcmc(theta0, 1000)
best_pos = sampler.flatchain[sampler.flatlnprobability.argmax()]
pos = emcee.utils.sample_ball(best_pos, best_pos/100., size=nwalkers)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 100)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 1000)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 1000)
```
