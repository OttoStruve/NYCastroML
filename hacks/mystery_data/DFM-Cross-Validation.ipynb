```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
data = np.loadtxt("data.csv", delimiter=",")
x, y, yerr = data.T
```

```python
pl.errorbar(x, y, yerr=yerr, fmt=".k");
```

```python
def fit(x, y, yerr, order=1):
    A = np.vander(x, order+1)
    C = np.diag(yerr**2)
    S = np.dot(A.T, np.linalg.solve(C, A))
    return np.linalg.solve(S, np.dot(A.T, np.linalg.solve(C, y))), np.linalg.inv(S)
```

```python
def model(w, x):
    A = np.vander(x, len(w))
    return np.dot(A, w)
```

```python
def crossvalid(K, x, y, yerr, order=1):
    ll = 0.0
    inds = np.random.randint(K, size=len(x))
    for k in range(K):
        m = inds == k
        w, c = fit(x[~m], y[~m], yerr[~m], order)
        ymod = model(w, x[m])
        ll += -0.5 * np.sum((y[m] - ymod) ** 2 / yerr[m]**2)
    return ll
```

```python
N = 100
ks = np.arange(1, 10)
cvl = np.empty((N, len(ks)))
for i in range(N):
    cvl[i, :] = [crossvalid(10, x, y, yerr, 2) for k in ks]
```

```python
mu = np.mean(cvl, axis=0)
std = np.std(cvl, axis=0)
```

```python
pl.errorbar(ks, mu, yerr=std, fmt=".k");
```

```python
def model2(w, x):
    return w[0] + w[1]*np.cosh(x/w[2])*x
```

```python
import scipy.optimize as op
def nll(p, x, y, yerr):
    return np.sum((y - model2(p, x)) ** 2 / yerr ** 2)

def fit2(x, y, yerr):
    return (op.minimize(nll, [6.1, 0.5, 8.0], method="L-BFGS-B", args=(x, y, yerr))).x, None
```

```python
def crossvalid2(K, x, y, yerr):
    ll = 0.0
    inds = np.random.randint(K, size=len(x))
    for k in range(K):
        m = inds == k
        w, c = fit2(x[~m], y[~m], yerr[~m])
        ymod = model2(w, x[m])
        ll += -0.5 * np.sum((y[m] - ymod) ** 2 / yerr[m]**2)
    return ll
```

```python
N = 100
ks = np.arange(1, 10)
cvl2 = np.empty(N)
for i in range(N):
    cvl2[i] = crossvalid2(10, x, y, yerr)
```

```python
mu_true = np.mean(cvl2)
std_true = np.std(cvl2)
print(mu_true, std_true)
```

```python
pl.fill_between(ks, (mu_true+std_true)*np.ones_like(ks), (mu_true-std_true)*np.ones_like(ks), color="k", alpha=0.3)
pl.errorbar(ks, mu, yerr=std, fmt=".k")
pl.plot(ks, mu_true*np.ones_like(ks), "k");
```
