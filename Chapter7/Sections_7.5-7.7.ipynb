# Section 7.5: Manifold Learning

__Non-linear dimensionality reduction__ assumes that not all dimensions of a particular dataset are interesting-- the "intrinsic variables" or data of interest therefore lie on some manifold embedded within the higher dimensional space. Representing the data with these nonlinear components is useful for reducing the size of the data set, and can be helpful in visualizing data. 

A really, really simple example:

```python
from IPython.core.display import Image 
Image(filename='NLDR.png')
```

Above is a three dimensional dataset comprised of the letter A at different scales and rotations. Because all of the elements are the letter A, we obviously care only about the scale and rotation of each element-- so these data can be represented in two dimensions (scale and rotation) rather than three. 

__Manifold learning techniques__ allow one to extract the intrinsic variables in the data while preserving information about the higher dimensional space, which is impossible using linear dimensionality reduction techniques like PCA. This is nicely illustrated in Figure 7.8:

```python
from IPython.core.display import Image 
Image(filename='fig_s_manifold_PCA_1.png')
```

If PCA is applied to the S-curve data at upper left, relationships between points in the higher dimensional (3D) space are lost-- the upper right panel shows mixing between points from unrelated parts of the S-curve when projected into two dimensions. 

The lower two panels show results from the two manifold learning techniques discussed in the chapter, __Local Linear Embedding (or LLE) and IsoMap__. There are many other manifold learning algorithms, some of which are discussed in the scikit-learn documentation:

## 7.5.1 Locally Linear Embedding (LLE)

LLE proceeds in two steps:

    1. derive a set of weights that best reconstruct a point from its k nearest neighbors
    2. hold these weights fixed and find a lower D data set that maintains the relationships between neighbors

The rows of the weight matrix therefore encode the local properties of the data set, while the entire weight matrix encodes the properties of the data set as a whole, as built up from the aggregated local knowledge. More specifically, LLE first minimizes

$$\mathcal{E}_1‎(W) ‎=‎ \mid X - WX \mid^2$$

or equivalently:
$$\mathcal{E}_1‎(W) ‎=‎ {\sum_{i=1}^{N}\mathbf{x}_{i}}\mid - {\sum_{i=1}^{N}W_{ij}\mathbf{x}_{j}}\mid^2$$

__Aside__: Do you know about Detexify? LaTeX math symbol classification FTW: http://detexify.kirelabs.org/classify.html

Where the diagonal elements of the weight matrix are zero, $W_{ii} = 0$, for all i, and elements of the weight matrix $W_{ij}$ = 0 for all j, unless j is one of the k nearest neighbors of i. The sum over all weights in a given row must be equal to 1. 

W is then held fixed, and a new matrix Y is found that minimizes 

$$\mathcal{E}_2(W) ‎=‎ \mid Y - WY \mid^2$$

so Y becomes the nonlinear representation of X.

```python
import numpy as np
from sklearn.manifold import LocallyLinearEmbedding 

X = np.random.normal(size=(1000, 2))

R = np.random.random((2,10))
X = np.dot(X, R) # 2D linear manifold in 10D space

k = 5 # num nearest neighbors
n = 2 # num dim in fit
lle = LocallyLinearEmbedding(k, n)
lle.fit(X)
proj = lle.transform(X)
```

__LLE was used to classify the Kepler Eclipsing Binary Catalogue__ by Matejivic et al. (http://arxiv.org/pdf/1204.2113v1.pdf). The authors used polyfit analytic models of phase-folded Kepler lightcurves, sampled at 1000 phase points, as their input dataset. They then projected the input data down to 3 dimensions, and then to 2 dimensions upon recognizing that the 3D projection was a fairly well-defined 2D distribution embedded in the 3D space. ## 7.5.2 IsoMap

__IsoMap__ is based on multidimensional scaling (MDS), but while MDS reconstructs data by computing a pairwise distance matrix between __all__ points in the data set, IsoMap takes a local approach and only computes distances between nearest neighbors. This local representation of the MDS distance matrix is the connected graph $G$, where $G_{ij}$ is the pairwise distance between points i and j __only if__ they are neighbors, and otherwise is zero. 

Cost wise, IsoMap is roughly comparable to LLE except for the computation of shortest paths-- there are a couple of clever workarounds for this last step which are included in scikit-learn.

Implementing IsoMap in python is almost identical to what we did above using LLE:

```python
from sklearn.manifold import Isomap # note M in IsoMap not capitalized here

iso = Isomap(k, n)
iso.fit(X)
proj = iso.transform(X)
```

## There's bad news and there's good news...

* Noisy and gappy data are problematic for manifold learning, causing parts of the manifold to be artifically connected to one another. 
    _Bright Side: Free thesis topic!_

* The solution obtained depends on neighbors used, and highly curved or non-uniformly populated spaces are problematic. There aren't really good recommendations on how to choose an optimal set of neighbors. _Bright Side: Free thesis topic!_

* There's no good way of estimating the dimensionality of the embedded manifold. _Bright Side: *crickets chirping*_

* Outliers are problematic in the same way that noisy or missing data is, causing parts of the manifold to be connected when they shouldn't be. _Bright side: See above_

* You don't really compress the data using manifold learning (unlike PCA, which creates a set of basis vectors). _Bright side: *shrug*_

# Section 7.6: ICA and (in passing) Projection Pursuit

__ICA__ reconstructs data from a linear combination of basis functions, where these components are taken as being statistically independent. Specifically, ICA seeks an unmixing matrix W that maximizes the non-Gaussianity of the underlying distributions. Different implementations of ICA use different definitions of non-Gaussianity.

```python
from sklearn.decomposition import FastICA

X = np.random.normal(size=(100, 2))

R = np.random.random((2, 5))
X = np.dot(X, R)

ica = FastICA(2)
ica.fit(X)
proj = ica.transform(X)

comp = ica.components_

sources = ica.fit_transform(X) # note that this differs from the code in the book; sources_ is deprecated, replaced by fit_transform
```

Note that the form and ordering of the underlying components found is dependent on technique and how many underlying components are assumed to be present. 

__Projection Pursuit__: it's a thing! Apparently. 

# Section 7.7: So which of these things should I use? 

__Accuracy__: Of the approaches presented in the chapter, only NMF can model the measured uncertainties, which gives it a leg up on some of the other options. However, non-linear methods may perform better overall, and in general the fewer constraints the less the reconstruction error. 

__Dimensionality Reduction: What Does It Mean?__ Generally the linear methods presented make more intuitive sense-- the PCA/NMF/ICA components shown in Figure 7.4 look more or less like the underlying physical signals they capture. This is not the case in the nonlinear methods, which are just maps between higher and lower dimensional spaces. 

__Scalability__: IsoMap and LLE are $\mathcal{O}(N^3)$ and $\mathcal{O}(N^2)$, respectively. 

__Simplicity__: the fact that you don't know the number of underlying components/dimensions makes all these methods a bit fiddly. 

For PCA/NMF/ICA, predictions can be made for out-of-sample data and the models therefore cross-validated. 
In the case of LLE and IsoMap, however, see problem above with not having a good recommendation for choosing the optimal set of neighbors-- you would have to add the test data to the original data and retrain, so cross validation isn't really possible with these methods.
