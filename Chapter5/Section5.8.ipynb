We want to compute an integral of the form

$$\mathrm{E}_p [f(\theta)] = \int f(\theta)\,p(\theta\,|\,x)\,\mathrm{d}\theta \quad.$$

We can evaluate this integral using Monte Carlo integration. If we can draw $N$ samples $\theta^{(n)} \sim q(\theta)$ then

$$\mathrm{E}_p [f(\theta)] \approx \frac{1}{N} \sum_{n=1}^N \frac{f(\theta^{(n)})\,p(\theta^{(n)}\,|\,x)}{q(\theta^{(n)})} \quad.$$

$q(\theta)$ needs to be *similar* to $p(\theta^{(n)}\,|\,x)$... best if we can actually sample from $p(\theta^{(n)}\,|\,x)$.

# Demo

We're going to fit a line to some data where the errorbars are underestimated.

First, the truth:

```python
truth = [1.1, -0.5, np.log(2.0)]
```

Generate some fake data.

```python
x = np.sort(5 - 10*np.random.rand(10))
y = truth[0] * x + truth[1]
yerr = 0.2 + 0.2 * np.random.rand(len(x))
y += np.sqrt(yerr**2 + np.exp(2 * truth[2])) * np.random.randn(len(y))
```

```python
pl.errorbar(x, y, yerr=yerr, fmt=".k", capsize=0);
```

Let's start by ignoring the fact that the errorbars are **wrong**:

```python
A = np.vander(x, 2)
C = np.diag(yerr**2)
S = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))
m, b = np.dot(S, np.dot(A.T, np.linalg.solve(C, y)))
display(Latex(r"$m_\mathrm{{LS}} = {0:.3f} \pm {1:.3f}$ ($m_\mathrm{{true}} = {2:.3f}$)".format(m, np.sqrt(S[0, 0]), truth[0])))
display(Latex(r"$b_\mathrm{{LS}} = {0:.3f} \pm {1:.3f}$ ($b_\mathrm{{true}} = {2:.3f}$)".format(b, np.sqrt(S[1, 1]), truth[1])));
```

The log-likelihood function is:

$$
\ln p(y\,|\,x,\,\sigma,\,m,\,b,\,s) = C-\frac{1}{2} \sum_{n=1}^N \left [ \frac{(y_n-m\,x_n-b)^2}{\sigma_n^2 + s^2} + \ln (\sigma_n^2 + s^2) \right ]
$$

```python
def lnlike(p):
    m, b, lns = p
    model = m * x + b
    invar = 1.0 / (yerr**2 + np.exp(2*lns))
    return -0.5 * np.sum((y-model)**2 * invar - np.log(invar))
```

We're going to use uniform priors on $m$, $b$ and $\ln s$:

```python
def lnprior(p):
    m, b, lns = p
    if -1 <= m <= 4 and -4 <= b <= 4 and -1 <= lns <= 4:
        return 0.0
    return -np.inf
```

Combine the prior and likelihood to get something proportional to the probability:

```python
def lnprob(p):
    return lnprior(p) + lnlike(p)
```

Here we'll just confirm that the log-probability of the "truth" is higher than another random point.

```python
print(lnprob(truth), lnprob(np.random.rand(3)))
```

## Monte Carlo and Importance Sampling

```python
N = 10000
```

We'll sample $N$ points from the (trivial) prior ($q(\theta) \propto 1$):

```python
samples = np.array([-1, -4, -1])[None, :] + np.array([5, 8, 5])[None, :] * np.random.rand(N, 3)
```

Compute the log-probability at each of these samples:

```python
lnprobs = map(lnprob, samples)
print(lnprobs)
```

The weight of each sample is proportional to $p(\theta^{(n)}\,|\,x) / q(\theta^{(n)}) = w^{(n)}$ and since none of our distirbutions are correctly normalized, our integral becomes:

$$\mathrm{E}_p [f(\theta)] \approx \left[ \sum_n w^{(n)} \right]^{-1} \sum_{n=1}^N f(\theta^{(n)})\,w^{(n)}$$

with weights computed as follows:

```python
weights = np.exp(lnprobs - np.max(lnprobs))
```

Here are the results:

```python
triangle.corner(samples, weights=weights, plot_datapoints=False,
                truths=truth, quantiles=[0.16, 0.5, 0.84],
                labels=["$m$", "$b$", "$\ln s$"]);
```

```python
for i, v in enumerate(["m", "b", "\ln s"]):
    norm = np.sum(weights)
    mu = np.sum(samples[:, i] * weights) / norm
    std = np.sqrt(np.sum((samples[:, i] - mu)**2 * weights) / norm)
    display(Latex(r"${0}_\mathrm{{MC}} = {1:.3f} \pm {2:.3f}$ (${0}_\mathrm{{true}} = {3:.3f}$)"
              .format(v, mu, std, truth[i])))
```

This looks good until we see that most of our samples are wasted:

```python
pl.hist(weights, 100, color="k", histtype="step")
pl.ylabel("number of samples")
pl.xlabel("sample weight");
```

## Metropolis Algorithm

MCMC draws samples from $p(\theta\,|\,x)$ directly while only needing us to compute it up to a constant.

**The algorithm**

1. start at some position $\theta^{(t)}$ in parameter space,
2. sample a new proposal position using some "proposal" distribution $\theta\prime \sim Q(\theta\prime\,|\,\theta^{(t)})$,
3. accept this proposal with probability
$$r = \min \left (1, \frac{Q(\theta^{(t)}\,|\,\theta\prime)}{Q(\theta\prime\,|\,\theta^{(t)})} \, \frac{p(\theta\prime\,|\,x)}{p(\theta^{(t)}\,|\,x)} \right ) \quad, \mathrm{and}$$
4. iterate forever.

The Metropolis algorithm assumes that $Q(\theta^{(t)}\,|\,\theta\prime) = Q(\theta\prime\,|\,\theta^{(t)})$ making the update even simpler. A simple pure Python implementation of a Metropolis step (assuming a Gaussian proposal) looks like:

```python
def metropolis_update(lnprobfn, cov, x0, lp0):
    q = np.random.multivariate_normal(x0, cov)
    lp1 = lnprobfn(q)
    diff = lp1 - lp0
    if diff >= 0.0 or np.exp(diff) >= np.random.rand():
        return q, lp1, True
    return x0, lp0, False
```

We'll use an axis-aligned Gaussian as the proposal:

```python
cov = np.diag([0.001, 0.001, 0.001])
```

And run for $N$ steps. I'll also be computing the acceptance fraction by keeping track of how many samples were saved.

```python
count = 0
chain = np.empty((N, 3))
chain[0] = np.array(truth)
lp0 = lnprob(chain[0])
for i in range(1, N):
    chain[i], lp0, accept = metropolis_update(lnprob, cov, chain[i-1], lp0)
    count += accept
print("The acceptance fraction is {0}".format(float(count) / N))
```

Finally, here are the results:

```python
triangle.corner(chain, truths=truth, quantiles=[0.16, 0.5, 0.84],
                labels=["$m$", "$b$", "$\ln s$"]);
```

```python
for i, v in enumerate(["m", "b", "\ln s"]):
    display(Latex(r"${0}_\mathrm{{MCMC}} = {1:.3f} \pm {2:.3f}$ (${0}_\mathrm{{true}} = {3:.3f}$)"
                  .format(v, np.mean(chain[:, i]), np.std(chain[:, i]), truth[i])))
```

## Using PyMC

```python
import pymc
```

```python
m = pymc.Uniform("m", -1, 4)
b = pymc.Uniform("b", -4, 4)
lns = pymc.Uniform("lns", -1, 4)

@pymc.deterministic(plot=False)
def invar(lns=lns):
    return 1.0 / (yerr**2 + np.exp(2*lns))

v = pymc.Normal("v", m * x + b, invar, observed=True, value=y)
model = dict(v=v, invar=invar, lns=lns, b=b, m=m)
S = pymc.MCMC(model)
```

```python
S.sample(N)
```

```python
pymc_chain = np.vstack((S.trace("m")[:], S.trace("b")[:], S.trace("lns")[:])).T
```

```python
triangle.corner(pymc_chain, truths=truth, quantiles=[0.16, 0.5, 0.84],
                labels=["$m$", "$b$", "$\ln s$"]);
```

```python
for i, v in enumerate(["m", "b", "\ln s"]):
    display(Latex(r"${0}_\mathrm{{PyMC}} = {1:.3f} \pm {2:.3f}$ (${0}_\mathrm{{true}} = {3:.3f}$)"
                  .format(v, np.mean(pymc_chain[:, i]), np.std(pymc_chain[:, i]), truth[i])))
```

## Using emcee

```python
import emcee
```

```python
ndim, nwalkers = 3, 10
p0 = [truth + 1e-1 * np.random.randn(ndim) for i in range(nwalkers)]

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)
```

```python
sampler.run_mcmc(p0, N / nwalkers);
```

```python
emcee_chain = sampler.flatchain
```

```python
triangle.corner(emcee_chain, truths=truth, quantiles=[0.16, 0.5, 0.84],
                labels=["$m$", "$b$", "$\ln s$"]);
```

```python
for i, v in enumerate(["m", "b", "\ln s"]):
    display(Latex(r"${0}_\mathrm{{emcee}} = {1:.3f} \pm {2:.3f}$ (${0}_\mathrm{{true}} = {3:.3f}$)"
                  .format(v, np.mean(emcee_chain[:, i]), np.std(pymc_chain[:, i]), truth[i])))
```
