# Fitting a model to data with outliers using MCMC

We are often faced with data with spurious outliers. For example, a light curve generated from some automated photometric pipeline that failed  the background subtraction one night. Your model should be able to account for these data -- outliers can contain information! Luckily, it is relatively easy to account for this within a Bayesian context -- each data point will be assigned a probability that it is drawn from your physical model vs. some other outlier model.

Below we consider a trivial, but illustrative, example. Note that this is described in a little more detail in section 5.6.7. Another good reference is David Hogg's [Fitting a model to data](http://arxiv.org/abs/1008.4686).

----

```python
import matplotlib.pyplot as plt
import numpy as np
import emcee
import triangle
from scipy.misc import logsumexp

np.random.seed(1234)
```

We take images of a perfectly sinusoidal variable star randomly over the course of a year. We have some automated pipeline that reduces the data, and spits out magnitudes for each observation. Some nights, the pipeline fails, but the way we estimate uncertainties on the magnitudes doesn't account for these failures, leading to outliers in the final data product (a light curve). We want to measure the amplitude and period of the star from the light curve.

```python
def sinusoid(t, amp, period, phase):
    return amp*np.sin(2*np.pi*t/period + phase)

def light_curve_model(p, t):
    """ A sinusoid + constant (magnitude) """
    amp, period, phase, const = p
    return sinusoid(t, amp, period, phase) + const

param_labels = ["Amp.", "Period [day]", "Phase [rad]", "Mean mag."]
```

First let's generate some fake data...

```python
ndata = 32
outlier_fraction = 0.1 # 10% outliers
true_params = (1.5, 112., 1.5, 14.)

time = np.random.uniform(0., 365., size=ndata)
mag = light_curve_model(true_params, time)

mag_err = np.random.uniform(0.1, 0.3, size=ndata)
mag += np.random.normal(0., mag_err)

outlier_idx = np.random.uniform(size=ndata) < outlier_fraction
mag[outlier_idx] += np.random.normal(0., np.random.uniform(3., 5., size=sum(outlier_idx)))

plt.figure(figsize=(12,4))
plt.errorbar(time, mag, mag_err, marker='o',
             linestyle='none', ecolor='#aaaaaa')
plt.xlim(0,365)
```

Let's start by fitting a model -- a sinusoid + constant -- to the data without accounting for the outliers.

Just by visual inspection of the light curve, we can set some reasonable priors on the parameters:

```python
def lnprior(p):
    amp,period,phase,const = p
    
    if amp < 1 or amp > 2:
        return -np.inf
    
    if period < 100 or period > 200:
        return -np.inf
    
    if phase < 0. or phase > 2*np.pi:
        return -np.inf
    
    if const < 12 or const > 16:
        return -np.inf
    
    return 0.

def lnlikelihood(p, t, data, err):
    amp,period,phase,const = p
    return -np.log(err) - 0.5*( (data - sinusoid(t, amp, period, phase) - const) / err )**2

def lnprob(p, t, data, err):
    return np.sum(lnprior(p) + lnlikelihood(p, t, data, err))
```

```python
ndim, nwalkers = 4, 32
p0 = np.zeros((nwalkers,ndim))
p0[:,0] = np.random.uniform(1, 2., size=nwalkers) # amp
p0[:,1] = np.random.uniform(100, 200., size=nwalkers) # period (days)
p0[:,2] = np.random.uniform(0., 2*np.pi, size=nwalkers) # phase (radians)
p0[:,3] = np.random.uniform(12., 16., size=nwalkers) # const. offset (mag)

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, 
                                args=(time,mag,mag_err))
```

```python
pos, prob, state = sampler.run_mcmc(p0, 1000)

best_pos = sampler.flatchain[sampler.flatlnprobability.argmax()]
pos = emcee.utils.sample_ball(best_pos, best_pos/100., size=nwalkers)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 100)

sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 1000)
```

```python
extents = [(0.5*truth,1.5*truth) for truth in true_params]
fig = triangle.corner(sampler.flatchain, 
                      labels=param_labels,
                      truths=true_params,
                      extents=extents)
```

Interesting -- we nailed the mean magnitude, but period, phase, and amplitude are biased.

Let's draw some samples from this posterior and plot the models over the data:

```python
nsamples = sampler.flatchain.shape[0]

plt.figure(figsize=(12,4))
plt.errorbar(time, mag, mag_err, marker='o',
             linestyle='none', ecolor='#aaaaaa')

t = np.linspace(min(time), max(time), 1000)
for ii in range(10):
    idx = np.random.randint(0, nsamples)
    params = sampler.flatchain[idx]
    model_mag = light_curve_model(params, t)
    plt.plot(t, model_mag, 
             marker=None, color='#f03b20', alpha=0.4)

plt.xlim(min(time), max(time))
```

What can we do about this? 

Mixture model! We add a new parameter for each data point -- the probability that it was instead drawn from some outlier distribution.

```python
def light_curve_model(p, t):
    """ A sinusoid + constant (magnitude) """
    amp, period, phase, const = p
    return sinusoid(t, amp, period, phase) + const

param_labels = ["Amp.", "Period [day]", "Phase [rad]", "Mean mag."]

def lnprior(p):
    amp = p[0]
    period = p[1]
    phase = p[2]
    const = p[3]
    outlier_prob = np.array(p[4:])
    
    if amp < 1 or amp > 2:
        return -np.inf
    
    if period < 100 or period > 200:
        return -np.inf
    
    if phase < 0. or phase > 2*np.pi:
        return -np.inf
    
    if const < 12 or const > 16:
        return -np.inf
    
    if np.any(outlier_prob > 1.) or np.any(outlier_prob < 0.):
        return -np.inf
    
    return 0.

def lnlikelihood(p, t, data, err):
    amp = p[0]
    period = p[1]
    phase = p[2]
    const = p[3]
    outlier_prob = np.array(p[4:])
    
    term1 = -np.log(err) - 0.5*( (data - sinusoid(t, amp, period, phase) - const) / err )**2
    
    outlier_err = 10.*np.median(err)
    term2 = -np.log(outlier_err) - 0.5*( (data - sinusoid(t, amp, period, phase) - const) / outlier_err )**2
    
    b = np.ones((2,len(outlier_prob)))
    b[0] = 1. - outlier_prob 
    b[1] = outlier_prob
    
    return logsumexp(np.vstack((term1, term2)), b=b, axis=0)
    
def lnprob(p, t, data, err):
    prior = lnprior(p)
    if np.isinf(prior):
        return -np.inf
    
    return prior + np.sum(lnlikelihood(p, t, data, err))
```

```python
ndim, nwalkers = 4+ndata, 256
p0 = np.zeros((nwalkers,ndim))
p0[:,0] = np.random.uniform(1, 2., size=nwalkers) # amp
p0[:,1] = np.random.uniform(100, 200., size=nwalkers) # period (days)
p0[:,2] = np.random.uniform(0., 2*np.pi, size=nwalkers) # phase (radians)
p0[:,3] = np.random.uniform(12., 16., size=nwalkers) # const. offset (mag)
p0[:,4:] = np.random.normal(0.5, 0.05, size=(nwalkers,ndata)) # outlier probabilty

sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, 
                                args=(time,mag,mag_err))
```

```python
pos, prob, state = sampler.run_mcmc(p0, 1000)
```

```python
best_pos = sampler.flatchain[sampler.flatlnprobability.argmax()]
pos = emcee.utils.sample_ball(best_pos, best_pos/100., size=nwalkers)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 100)

sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, 1000)
```
